```python
# load data from CSV
data = pd.read_csv('MBTA_Line_and_Stop.csv')
```

```python
# Check table for any null values to ensure clean up process is complete
data.isnull().values.any()
```

```python
data.head()
```

```python
df_weekend = data.loc[data['day_type_name'] != "weekday"]
df_weekday = data.loc[data['day_type_name'] == "weekday"]
```

```python
df_orange_green = df_weekday[df_weekday.route_id.isin(['Green', 'Orange'])]
df_orange_green.shape
```

```python
brush = alt.selection(type='interval', resolve='global')

base = alt.Chart(df_orange_green, title="Average Number of Passengers Onboarding vs Offboarding the Orange and Green Line Trains on Weekdays").mark_point().encode(
    x=alt.X('average_ons', axis=alt.Axis(title='Average Number of Daily Onboarding Passengers')),
    y=alt.Y('average_offs', axis=alt.Axis(title='Average Number of Daily Offboarding Passengers')),
    size=alt.condition(
        alt.FieldOneOfPredicate(field='stop_name', oneOf=['Ruggles', 'Northeastern']),
        alt.value(250),     
        alt.value(10)     
    ),
    color=alt.condition(brush, 'route_name', alt.ColorValue('gray')),
    tooltip='stop_name',
).add_selection(
    brush
).properties(
    width=1000,
    height=1000
)

base2 = alt.Chart(df_orange_green, title="Average Flow of Passengers for Orange and Green Line Routes on Weekdays Across Different Time Period").mark_line(point=True).encode(
    x=alt.X('time_period_name', axis=alt.Axis(title="Time Period")),
    y=alt.Y('mean(average_flow)', axis=alt.Axis(title="Average Flow of Passengers Onboarding Route")),
    color=alt.condition(brush, 'route_name', alt.ColorValue('gray')),
).add_selection(
    brush
).properties(
    width=1000,
    height=1000
)

base | base2
```
